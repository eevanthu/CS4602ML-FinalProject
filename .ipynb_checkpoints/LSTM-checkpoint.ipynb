```python
import pandas as pd
import numpy as np
import torch
import torch.nn as nn
from torch.utils.data import Dataset, DataLoader
from sklearn.preprocessing import MinMaxScaler
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("2330.csv")

#transfowm data format
df["Date"] = pd.to_datetime(df["Date"], format="%Y-%m-%d")

split_date = pd.to_datetime("2025/1/1")

df_before = df[df["Date"] < split_date]
df_after = df[df["Date"] >= split_date]

df_before.to_csv("train.csv", index=False)
df_after.to_csv("test.csv", index=False)
```

```python
device = torch.device('cuda')

df_train = pd.read_csv('train.csv')
df_test = pd.read_csv('test.csv')
# features used to train
features = ['Close', 'Turnover', 'High', 'Low', 'Transaction']
target_feature = 'Close'
#deal with missing data
df_train[features] = df_train[features].apply(pd.to_numeric, errors='coerce')
df_test[features] = df_test[features].apply(pd.to_numeric, errors='coerce')
df_train.ffill(inplace=True)
df_test.ffill(inplace=True)
df_train.bfill(inplace=True)
df_test.bfill(inplace=True)
print(df_train.head(5))
#transform df data to numpy
training_features = df_train[features]



#scaler features to range (0,1)
scaler = MinMaxScaler(feature_range=(0, 1))
scaled_training_data = scaler.fit_transform(training_features)
```

```python
#hyper parameter
sequence_length = 40
batch_size = 64
epochs = 50
lr = 0.001
```

```python
#split data
def create_sequences(data, sequence_length):
  X, y = [], []
  for i in range(len(data) - sequence_length):
      X.append(data[i:(i + sequence_length)])
      y.append(data[i + sequence_length])
  return np.array(X), np.array(y)


X_train, y_train = create_sequences(scaled_training_data, sequence_length)
#y only need feature close
target_col_index = features.index(target_feature)
y_train = y_train[:, target_col_index]

# change data to PyTorch Tensors
X_train = torch.tensor(X_train, dtype=torch.float32)
y_train = torch.tensor(y_train, dtype=torch.float32).reshape(-1, 1)

print(f"X_train shape: {X_train.shape}")
print(f"y_train shape: {y_train.shape}")
```

```python
#pipeline
class StockDataset(Dataset):
    def __init__(self, X, y):
        self.X = X
        self.y = y

    def __len__(self):
        return len(self.X)

    def __getitem__(self, idx):
        return self.X[idx], self.y[idx]

train_dataset = StockDataset(X_train, y_train)
train_loader = DataLoader(train_dataset, batch_size=batch_size, shuffle=True)
```

```python
# design model
class StockLSTM(nn.Module):
    # design the model's building blocks
    def __init__(self, input_size, hidden_dim, num_layers, output_size):
        super(StockLSTM, self).__init__()
        self.hidden_dim = hidden_dim
        self.num_layers = num_layers

        self.lstm = nn.LSTM(input_size, hidden_dim, num_layers, batch_first=True,dropout=0) # LSTM layer

        self.linear = nn.Linear(hidden_dim, output_size) # linear layer

    def forward(self, x):
        #initial hidden state and cell state for LSTM
        h0 = torch.zeros(self.num_layers, x.size(0), self.hidden_dim).to(device)
        c0 = torch.zeros(self.num_layers, x.size(0), self.hidden_dim).to(device)

        out, _ = self.lstm(x, (h0, c0))

        out = self.linear(out[:, -1, :])
        return out

#parameter
input_size = len(features)
hidden_dim = 64
num_layers = 2
output_size = 1

model = StockLSTM(input_size, hidden_dim, num_layers, output_size).to(device)
print(model)
```

```python
# use MSE loss function
criterion = nn.MSELoss()

optimizer = torch.optim.Adam(model.parameters(), lr=lr)

model.train()
for epoch in range(epochs):
    for X_batch, y_batch in train_loader:
        X_batch, y_batch = X_batch.to(device), y_batch.to(device)# move data to GPU
        # calculate output
        outputs = model(X_batch)
        #calculate loss
        loss = criterion(outputs, y_batch)
        #clear previous gradient
        optimizer.zero_grad()
        #back propogation
        loss.backward()
        #optimize
        optimizer.step()

    if (epoch + 1) % 5 == 0:
        print(f'Epoch [{epoch+1}/{epochs}], Loss: {loss.item():.6f}')
```

```python
#visualize
model.eval()
# prepare the previous 60 day to for the first day in test data
last_60_days_train = df_train[features].tail(sequence_length)
#transform date format for plotting
df_test['Date'] = pd.to_datetime(df_test['Date'])
#extract columns used in training
test_features = df_test[features]
#merge test data and previous 60 day
input_data = pd.concat([last_60_days_train, test_features], axis=0)

input_data = scaler.transform(input_data)

# split data for question and answer
X_test, _ = create_sequences(input_data, sequence_length)
X_test = torch.tensor(X_test, dtype=torch.float32).to(device)
y_test_actual = df_test[target_feature]

results_list = []
#test
with torch.no_grad():
    for i in range(len(X_test)):
        question = X_test[i].unsqueeze(0)
        prediction_scaled = model(question)

        #convert scaler value into original value
        dummy_array = np.zeros((1, len(features)))# create an array to conform the shape of scaler
        target_col_index = features.index(target_feature)
        dummy_array[0, target_col_index] = prediction_scaled.cpu().item()
        prediction = scaler.inverse_transform(dummy_array)[0, target_col_index]

        result_row = {
            'Date': df_test['Date'][i],
            'Predicted_Close': prediction,
            'Actual_Close': y_test_actual[i],
            'Error': prediction - y_test_actual[i]
        }
        #save data
        results_list.append(result_row)

# save as csv file
results = pd.DataFrame(results_list)
output_csv_path = 'prediction_results.csv'
results.to_csv(output_csv_path, index=False, float_format='%.0f')



# plot
predictions = results['Predicted_Close']
plot_dates = results['Date']
y_to_plot = results['Actual_Close']

plt.figure(figsize=(14, 7))
plt.plot(plot_dates, y_to_plot, color='blue', label='Actual Price')
plt.plot(plot_dates, predictions, color='red', label='Predicted Price')
plt.title('Stock Price Prediction')
plt.xlabel('Date')
plt.ylabel('Stock Price')
plt.xticks(rotation=45)
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.savefig('stock_price_prediction.png')  # dpi=300 for high quality

plt.show()

rmse = np.sqrt(np.mean((predictions - y_to_plot)**2))
print(f'RMSE: {rmse:.4f}')
	
```
